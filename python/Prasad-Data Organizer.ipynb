```python
def splitRow(table):
    newframe=pd.DataFrame(columns=table.columns.values)
    size=0
    for i in range(len(table)):
        row = table.iloc[i]
        if '\n' in str(row['Street number']):
            data = row['Street number'].split('\n')
            for d in data:
                newrow = row
                newrow['Street number'] = d
                newframe.loc[size]=newrow
                size+=1
        else:
            newframe.loc[size]=row
            size+=1
    return newframe
```

```python
def splitRowTransfer(table):
    newframe=pd.DataFrame(columns=table.columns.values)
    size=0
    for i in range(len(table)):
        row = table.iloc[i]
        if '\n' in str(row['Previous Transfer of deed People Involved']):
            data = row['Previous Transfer of deed People Involved'].split('\n')
            for d in data:
                newrow = row
                newrow['Previous Transfer of deed People Involved'] = d
                newframe.loc[size]=newrow
                size+=1
        else:
            newframe.loc[size]=row
            size+=1
    return newframe
```

```python
def splitRowDecisions(table):
    newframe=pd.DataFrame(columns=table.columns.values)
    size=0
    for i in range(len(table)):
        row = table.iloc[i]
        if '\n' in str(row['Decision People Involved']):
            data = row['Decision People Involved'].split('\n')
            for d in data:
                newrow = row
                newrow['Decision People Involved'] = d
                newframe.loc[size]=newrow
                size+=1
        else:
            newframe.loc[size]=row
            size+=1
    return newframe
```

```python
decisions= splitRowDecisions(people)
```

```python
def combined(df1,df2):
    newdf=pd.DataFrame(columns=['People'])
    size=0
    for i in range(len(df1)):
        row=df1.iloc[i]
        data = row['Previous Transfer of deed People Involved']
        newdf.loc[size]=data
        size+=1
    for y in range(len(df2)):
        row=df2.iloc[y]
        data = row['Decision People Involved']
        newdf.loc[size]=data
        size+=1
    return newdf

```
